```python
sns.set_style('darkgrid')
f, axes = plt.subplots(2, 2, figsize=(15,15))
k = sns.kdeplot (movies.BudgetMillions, movies.AudienceRating, ax=axes[0,0])
k1 = sns.kdeplot (movies.BudgetMillions, movies.CriticRating, ax=axes[0,1])
k.set(xlim=(-20,160))
k1.set(xlim=(-20,160))
z = sns.violinplot(data=movies, x='Genre', y='CriticRating', ax=axes[1,0])
k2 = sns.kdeplot(movies.CriticRating, movies.AudienceRating, shade = True, shade_lowest=False, cmap='Reds', ax=axes[1,1])
k3 = sns.kdeplot(movies.CriticRating, movies.AudienceRating, shade = True, cmap='Reds', ax=axes[1,1])
```

```python
# Styling dashboards
```

```python

sns.set_style('white', {"axes.facecolor": 'black'}) #white, whitegrid
f, axes = plt.subplots(2, 2, figsize=(15,15))

#plot [0,0]
k = sns.kdeplot (movies.BudgetMillions, movies.AudienceRating, \
                 cmap='cool', ax=axes[0,0])
kb = sns.kdeplot (movies.BudgetMillions, movies.AudienceRating, \
                 shade=True, shade_lowest=True, cmap='inferno', ax=axes[0,0])

#plot [0,1]
k1 = sns.kdeplot (movies.BudgetMillions, movies.CriticRating, \
                  cmap='cool', ax=axes[0,1])
k1b = sns.kdeplot (movies.BudgetMillions, movies.CriticRating, \
                 shade=True, shade_lowest=True, cmap='inferno', ax=axes[0,1])

#plot [1,0]
z = sns.violinplot(data=movies, x='Year', y='BudgetMillions', ax=axes[1,0], \
                  palette='YlOrRd')

#plot [1,1]
k2 = sns.kdeplot(movies.CriticRating, movies.AudienceRating, \
                 shade = True, shade_lowest=False, cmap='Blues_r', ax=axes[1,1])
k3 = sns.kdeplot(movies.CriticRating, movies.AudienceRating, \
                 cmap='gist_gray_r', ax=axes[1,1])

k.set(xlim=(-20,160))
k1.set(xlim=(-20,160))
```

```python
# Thematic Touchup
```

```python
list1 = list()
mylabels = list()

for gen in movies.Genre.cat.categories:
    list1.append(movies[movies.Genre == gen].BudgetMillions)
    mylabels.append(gen)

sns.set_style("whitegrid")
fig, ax = plt.subplots()
fig.set_size_inches(11.7, 8.27)
h= plt.hist(list1, bins=30, stacked=True, rwidth=1, label=mylabels)
plt.title("Movie Budget Distribution", fontsize=35, \
         color='DarkBlue')
plt.ylabel('Number of Movies', fontsize=25, color='Red')
plt.xlabel('Budget', fontsize=25, color='Green')
plt.yticks(fontsize=20)
plt.xticks(fontsize=20)
plt.legend(prop={'size':20}, shadow=True)
plt.show()
```
